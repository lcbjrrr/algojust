# Descriptive Statistics

Descriptive statistics are crucial for summarizing and presenting the main features of a dataset in a clear and understandable way. They provide a foundational understanding of the data's characteristics, such as central tendency (mean, median, mode), variability (range, standard deviation, variance), and distribution (skewness, kurtosis). By simplifying large amounts of data into manageable summaries, descriptive statistics enable analysts to quickly grasp key insights, identify patterns, and detect potential issues, which is essential for informed decision-making and for guiding further, more complex analyses



> read the file 'grades - ok.csv' and show the dataframe



```python
import pandas as pd

grades = pd.read_csv('grades - ok.csv')
display(grades)
```

### Frequency Calculation and Historgrams

Frequency calculation involves counting how often each distinct value or category appears in a dataset, providing insights into the distribution and common occurrences of specific data points

Let's calculate the frequency of each score in the 'Score' column, prints these frequencies, and then displays them as a bar chart.



> calculate the frequency of the Score column



```python
# Calculate the frequency of each score
score_counts = grades['Score'].value_counts()
print(score_counts)
```

A histogram visualizes the distribution of a continuous variable using bars for frequency within data ranges (bins), typically with touching bars.



> Plot a bar chart for those



```python
import matplotlib.pyplot as plt

score_counts.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Frequency of Scores')
plt.xlabel('Score')
plt.ylabel('Frequency')
plt.show()
```

*The scores A, C, and E are the most frequent, each appearing 3 times (the tallest bars). Score B appears 2 times, and score D is the least frequent, appearing only once (the shortest bar)*.

 Let's use the seaborn library to generate a histogram of the 'Grade' column from the grades DataFrame, using custom bins to define the grade ranges.



> Plot the histogram for the Grade column considering as bins = 0,59,69,79,89,100



```python
import seaborn as sns
import matplotlib.pyplot as plt

# Define the bins
bins = [0, 59, 69, 79, 89, 100]

# Create the histogram
plt.figure(figsize=(10, 6))
sns.histplot(grades['Grade'], bins=bins, kde=False, color='skyblue', edgecolor='black')

# Add title and labels
plt.title('Grade Distribution')
plt.xlabel('Grade')
plt.ylabel('Frequency')
plt.xticks(bins)

plt.show()
```

*Once again, the scores A, C, and E are the most frequent, each appearing 3 times (the tallest bars). Score B appears 2 times, and score D is the least frequent, appearing only once (the shortest bar).*

## Quartiles and Boxplot

Quartiles and boxplots are important because they provide a concise visual and statistical summary of data distribution, central tendency, spread, and the presence of outliers, facilitating quick understanding and comparison of datasets.

Quartiles divide a dataset into four equal parts, helping to understand the spread and central tendency of the data. The first quartile (Q1) marks the 25th percentile, the second quartile (Q2) is the median (50th percentile), and the third quartile (Q3) is the 75th percentile.



> Using describe, calculate the quartiles for the Grade column



```python
# Calculate descriptive statistics including quartiles using describe()
grade_stats = grades['Grade'].describe()
display(grade_stats)
```

*Based on the describe() output, the quartiles for the Grade column are:*

*First Quartile (25%): **58.5**.*
*Median (50%): **75.0**.*
*Third Quartile (75%): **86.75**.*

A boxplot (or box-and-whisker plot) is a graphical representation that summarizes these quartiles, along with the minimum and maximum values, and identifies potential outliers, providing a quick visual summary of the data distribution and skewness.



> Create a boxplot for the Grade column



```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a boxplot for the 'Grade' column
plt.figure(figsize=(8, 6))
sns.boxplot(y=grades['Grade'])
plt.title('Boxplot of Grades')
plt.ylabel('Grade')
plt.show()
```

*Based on the boxplot for the 'Grade' column, I can describe the distribution as follows:*

*Median: The line inside the box is at 75.0, indicating that half of the grades are above and half are below this value.*

*Interquartile Range (IQR): The box spans from 58.5 (Q1) to 86.75 (Q3). This means the middle 50% of the grades fall within this range.*

*Outliers: There do not appear to be any individual data points marked as outliers outside the whiskers, suggesting all grades fall within the typical range for this dataset.*

With a boxplot we can visualize the distribution of 'Grade' for each 'Course' in the grades DataFrame



> Break it down per Course



```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.boxplot(x='Course', y='Grade', data=grades)
plt.title('Grade Distribution per Course')
plt.xlabel('Course')
plt.ylabel('Grade')
plt.show()
```

This box plot compares Grade distributions for three courses: ADM, LAW, and ECO. The ECO course has the most consistent grades, with a very tight cluster (small range, narrow box) around a median of $\sim 80$. The LAW course shows the highest grades (median $\sim 87$) and the greatest variability (widest box and overall range), spanning from $\sim 38$ to $\sim 98$. The ADM course has the lowest median grade ($\sim 68$) and a moderate spread compared to the others, indicating a central tendency toward mid-range scores. Overall, the data suggests high consistency in ECO, high performance but high variance in LAW, and moderate performance and variance in ADM.
