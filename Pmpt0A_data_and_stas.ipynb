# Data Analysis

Descriptive Data Analysis are the fundamental first steps in any data analysis project, serving as crucial tools for understanding, cleaning, and preparing data. They allow for the initial exploration of a dataset's characteristics, including its distributions, central tendencies, and variability. Through techniques such as statistical summaries and visualizations, EDA helps identify inherent patterns, relationships between variables, and anomalies or outliers that might otherwise go unnoticed, all of which are vital for ensuring data quality and reliability.

Beyond initial data comprehension, EDA plays a critical role in guiding subsequent analytical steps. Insights gained from this process inform data cleaning and preprocessing efforts, assist in the generation of hypotheses, and aid in feature engineering, ultimately leading to more robust models. By providing a clear picture of the data's structure and characteristics, descriptive analysis and EDA empower analysts to select appropriate statistical or machine learning models and effectively communicate key findings to both technical and non-technical stakeholders.

## Aggregation and Consolidation

Data aggregation and consolidation are crucial for transforming raw, detailed datasets into more meaningful and manageable forms. By summarizing and combining data from various sources or at different granularities, these processes help in identifying overarching trends, patterns, and anomalies that might be obscured in granular data. This simplification facilitates more efficient analysis, enhances decision-making by providing a higher-level view, and is essential for tasks like reporting, performance monitoring, and strategic planning. Ultimately, aggregation and consolidation enable analysts to extract actionable insights and communicate complex information effectively to stakeholders.

For example, you can group the grades DataFrame by 'Course', calculates the average 'Grade' for each course, and then displays the result.



> create an aggregated table grouping by course and calculate the average grade per course



```python
# Group by 'Course' and calculate the average 'Grade'
average_grade_course = grades.groupby('Course')['Grade'].mean()
display(average_grade_course)
```

## Comparision (Bar Chart)

***Bar charts*** are highly effective for **comparing data** across different categories or groups. They utilize rectangular bars of varying lengths, where each bar represents a specific category and its length corresponds to the value of the quantitative variable being measured. The visual distinction in bar lengths makes it easy to quickly identify which categories have higher or lower values, allowing for immediate comparisons. For instance, in a bar chart showing average grades by course, you can instantly see which course has the highest or lowest average grade by simply looking at the height of the bars. This visual clarity is particularly useful for highlighting differences, identifying trends, and communicating insights about categorical data.

You can generate a bar chart to visualize the average_grade_course data, making it easy to compare average grades across different courses.



> show a bar chart comparing those



```python
import matplotlib.pyplot as plt

# Plot the average grades as a bar chart
average_grade_course.plot(kind='bar', title='Average Grade by Course', color=['skyblue', 'lightgreen', 'salmon'])
plt.xlabel('Course')
plt.ylabel('Average Grade')
plt.show()
```

*The ECO course has the highest average grade at 79.5, followed by LAW at 72.4, and ADM at 67.2.*

## Proportion (Pie Chart)

***Pie charts*** are excellent for **visualizing proportions** and parts of a whole. Each slice of the pie represents a category, and the size of the slice corresponds to the proportion or percentage that category contributes to the total. This visual representation makes it easy to quickly grasp the relative contribution of each component to the overall dataset.

For example, let's group the grades DataFrame by 'Course', counts the number of students in each course, and then visualizes these proportions using a pie chart.



> now aggregate again by course and count the number of students per course



```python
# Group by 'Course' and count the number of students
student_counts = grades.groupby('Course')['Student'].count()
display(student_counts)
```



> show a pie chart



```python
import matplotlib.pyplot as plt

# Plot the distribution of students per course as a pie chart
student_counts.plot.pie(autopct='%1.1f%%', title='Proportion of Students per Course')
plt.ylabel('')  # Hide the y-label for a cleaner view
plt.show()
```

*The ADM and LAW courses have the most students with 5 each, while ECO has 2 students.*

# Quartiles and Boxplot

Quartiles and boxplots are important because they provide a concise visual and statistical summary of data distribution, central tendency, spread, and the presence of outliers, facilitating quick understanding and comparison of datasets.

Quartiles divide a dataset into four equal parts, helping to understand the spread and central tendency of the data. The first quartile (Q1) marks the 25th percentile, the second quartile (Q2) is the median (50th percentile), and the third quartile (Q3) is the 75th percentile.

A boxplot (or box-and-whisker plot) is a graphical representation that summarizes these quartiles, along with the minimum and maximum values, and identifies potential outliers, providing a quick visual summary of the data distribution and skewness.

With a boxplot we can visualize the distribution of 'Grade' for each 'Course' in the grades DataFrame

This box plot compares Grade distributions for three courses: ADM, LAW, and ECO. The ECO course has the most consistent grades, with a very tight cluster (small range, narrow box) around a median of $\sim 80$. The LAW course shows the highest grades (median $\sim 87$) and the greatest variability (widest box and overall range), spanning from $\sim 38$ to $\sim 98$. The ADM course has the lowest median grade ($\sim 68$) and a moderate spread compared to the others, indicating a central tendency toward mid-range scores. Overall, the data suggests high consistency in ECO, high performance but high variance in LAW, and moderate performance and variance in ADM.
